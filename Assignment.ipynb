```python
import math


def sum_of_squares(n):
    a = int(math.sqrt(n))
    for i in range(0, a + 1):
        for j in range(0, a + 1):
            if (i**2) + (j**2) == n:
                return True
    return False


n = int(input("Enter a number: "))
print(sum_of_squares(n))
```

```python
def compute_grade(grade):
    if grade > 0.8:
        return 'A'
    elif grade > 0.7:
        return 'B'
    elif grade > 0.6:
        return 'C'
    elif grade > 0.5:
        return 'D'
    elif grade > 0.4:
        return 'E'
    else:
        return 'F'


try:
    grade = float(input("Enter a grade:"))
    if 0.0 <= grade <= 1.0:
        print(compute_grade(grade))
    else:
        print("Invalid entry")
except ValueError:
    print("Not a valid type")
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact = fact * i
    print(fact)


def fibonacci(n):
    fib1 = 0
    fib2 = 1
    fib3 = fib1 + fib2
    print(fib1)
    print(fib2)
    for i in range(2, n):
        print(fib3)
        fib1 = fib2
        fib2 = fib3
        fib3 = fib1 + fib2


try:
    n = int(input("Enter a number:"))
    print("Factorial:")
    factorial(n)
    print("Fibonacci")
    fibonacci(n)
except ValueError:
    print("Not an integer.")
```

```python
def hcf(x, y):
    while (y):
        x, y = y, x % y
    return x


def lcm(x, y):
    lcm = (x * y) // hcf(x, y)
    return lcm


a = int(input("Enter a value for a:"))
b = int(input("Enter a value for b:"))

print(hcf(a, b))
print(lcm(a, b))
```

# 23

```python
def horner(poly, n, x):
    result = poly[0]
    for i in range(1, n):
        result = result * x + poly[i]
    print("The value of polynomial at x = " + str(x) + " is " + str(result))


a0 = int(input("Enter a value for a0:"))
a1 = int(input("Enter a value for a1:"))
a2 = int(input("Enter a value for a2:"))
a3 = int(input("Enter a value for a3:"))
a4 = int(input("Enter a value for a4:"))

poly = [a0, a1, a2, a3, a4]
x = int(input("Enter the value for x:"))
n = len(poly)
horner(poly, n, x)
```

```python
def is_prime(a):
    flag = 0
    for i in range(2, a):
        if a % i == 0:
            flag = 1
            break
    if flag == 1:
        return False
    else:
        return True


lower = int(input("Enter the lower value:"))
upper = int(input("Enter the upper value:"))

for i in range(lower, upper + 1):
    if is_prime(i):
        print(i)
```
