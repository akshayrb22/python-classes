# Classes and objects

```python
class student:
    def __init__(self, name, usn):
        self.name = name
        self.usn = usn

    def change(self, name):
        self.name = name


s = student("sanjay", 77)
print(s.name)
print(s.usn)
s.change("akshay")
print(s.name)
```

```python
s.usn = 43
print(s.usn)
```
