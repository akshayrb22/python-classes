```python
from math import pi


class Circle:
    def __init__(self, r):
        self.r = r

    def area(self, a):
        return pi*a*a

    def perimeter(self):
        return 2*pi*r


c = Circle(7)
#r = int(input("enter the radius "))
print("Area = ", c.area(5))
print("Perimeter = ", c.perimeter())
```

```python
class Bank:
    def __init__(self,name,acc_no,bal):
        self.name=name
        self.acc_no=acc_no
        self.bal=bal
        
    def deposit(self,amount):
        self.bal = self.bal + amount
        print("After depositing balance = ",self.bal)
        
    def withdrawal(self,amount):
        self.bal = self.bal - amount
        print("After withdrawing balance = ",self.bal)
        
b = Bank("sanjay",123456,10000)
b.deposit(5000)
b.withdrawal(4000)
```
