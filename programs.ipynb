# Palindrome

```python
s=input("enter a string: ")
rev=s[::-1]
if s==rev:
    print("palindrome")
else:
    print("not a palindrome")
```

# String slicing

```python
s="Good evening"
print(s[4])
print(s[-1])
print(s[4:8])
print(s[4:])
print(s[:4])
print(s[-7:-3])
print(s[4:8:1])
print(s[4:8:2])
print(s[8:4:-1])
print(s[4::-1])
print(s[:4:-1])
```
